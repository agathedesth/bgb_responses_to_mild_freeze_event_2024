```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.font_manager import FontProperties
import math
import statistics
from numpy import *
from matplotlib.pyplot import *
from matplotlib.patches import FancyBboxPatch, FancyArrowPatch


# Changing font to stix; setting specialized math font properties as directly as possible
rcParams['mathtext.fontset'] = 'custom'
rcParams['mathtext.it'] = 'STIXGeneral:italic'
rcParams['mathtext.bf'] = 'STIXGeneral:italic:bold'


###################################################################################################################################################################################################################################################
                                              
    
                                                # FOR THE DATA OF 2022

                                            # TO IMPORT THE DATA FROM CSV
    
# To import data: (Note, I took out the index -- that was making it treat the first column differently)
data = pd.read_csv('/Users/agathe/Library/CloudStorage/OneDrive-UniversityofNewOrleans/Foster-Martinez_Coastal_Interactions_Lab_shared/Agathe_Desthomas_Progress_Reports/Python/Total_2022_2023_Combined/BGB_Per_Site_December_2022.csv', header=0)
df1 = pd.DataFrame(data)



                                       # TO CHANGE THE FONT STYLE OF THE PLOT

# For the plot font:
font = FontProperties()
font.set_family('serif')
font.set_name('Times New Roman')



                               # TO GET RID OF THE NaN VALUES (EMPTY CELLS) FROM THE CSV FILE

# To get rid of the NaN values (empty cells in csv/excel) and replace them by 0.000:
df = df1.fillna(0.000)



                                        # TO GET THE SUM OF EACH CORE

# SITE A - To sum core by core:
core_1_site_A = df.iloc[0:,1]
sum_core_1_site_A = sum(core_1_site_A)
print("The sum of the site A core 1 is", sum_core_1_site_A, "g")

core_2_site_A = df.iloc[0:,2]
sum_core_2_site_A = sum(core_2_site_A)
print("The sum of the site A core 2 is", sum_core_2_site_A, "g")

core_3_site_A = df.iloc[0:,3]
sum_core_3_site_A = sum(core_3_site_A)
print("The sum of the site A of core 3 is", sum_core_3_site_A, "g")



# SITE B - To sum core by core:
core_1_site_B = df.iloc[0:,4]
sum_core_1_site_B = sum(core_1_site_B)
print("The sum of the site B core 1 is", sum_core_1_site_B, "g")

core_2_site_B = df.iloc[0:,5]
sum_core_2_site_B = sum(core_2_site_B)
print("The sum of the site B core 2 is", sum_core_2_site_B, "g")

core_3_site_B = df.iloc[0:,6]
sum_core_3_site_B = sum(core_3_site_B)
print("The sum of the site B of core 3 is", sum_core_3_site_B, "g")

core_4_site_B = df.iloc[0:,7]
sum_core_4_site_B = sum(core_4_site_B)
print("The sum of the site B core 4 is:", sum_core_4_site_B, "g")



# SITE C - To sum core by core:
core_1_site_C = df.iloc[0:,8]
sum_core_1_site_C = sum(core_1_site_C)
print("The sum of the site C core 1 is", sum_core_1_site_C, "g")

core_2_site_C = df.iloc[0:,9]
sum_core_2_site_C = sum(core_2_site_C)
print("The sum of the site C core 2 is", sum_core_2_site_C, "g")

core_3_site_C = df.iloc[0:,10]
sum_core_3_site_C = sum(core_3_site_C)
print("The sum of the site C of core 3 is", sum_core_3_site_C, "g")

core_4_site_C = df.iloc[0:,11]
sum_core_4_site_C = sum(core_4_site_C)
print("The sum of the site C core 4 is", sum_core_4_site_C, "g")




                                    # TO FIND THE AREA OF THE PLASTIC AND METAL CORES

# To find the area of the plastic core when core was entirely filled (0.30m):
diameter_plastic_core = 0.10
radius_plastic_core = diameter_plastic_core/2
length_plastic_core = 0.30
area_plastic_core = (math.pi*(radius_plastic_core)**2)



# To find the area of the plastic core when core was entirely filled (0.30m):
diameter_metal_core = 0.072
radius_metal_core = diameter_metal_core/2
length_metal_core = 0.30
area_metal_core = (math.pi*(radius_metal_core)**2)




                           # TO FIND THE BGB BY TAKING THE SUM AND DIVIDING IT BY THE AREA OF THE CORE

# SITE A - To find the BGB with the area for each core:
BGB_core_1_site_A = sum_core_1_site_A/area_plastic_core
print("The BGB for core 1 site A is:", BGB_core_1_site_A, "g m^-2")

BGB_core_3_site_A = sum_core_3_site_A/area_plastic_core
print("The BGB for core 3 site A is:", BGB_core_3_site_A, "g m^-2")

BGB_core_2_site_A = sum_core_2_site_A/area_plastic_core
print("The BGB for core 2 site A is:", BGB_core_2_site_A, "g m^-2")


# SITE B - To find the BGB with the area for each core:
BGB_core_1_site_B = sum_core_1_site_B/area_plastic_core
print("The BGB for core 1 site B is:", BGB_core_1_site_B, "g m^-2")

BGB_core_2_site_B = sum_core_2_site_B/area_plastic_core
print("The BGB for core 2 site B is:", BGB_core_2_site_B, "g m^-2")

BGB_core_3_site_B = sum_core_3_site_B/area_plastic_core
print("The BGB for core 3 site B is:", BGB_core_3_site_B, "g m^-2")

# Area of metal core for core 4:
BGB_core_4_site_B = sum_core_4_site_B/area_metal_core
print("The BGB for core 4 site B is:", BGB_core_4_site_B, "g m^-2")


# SITE C - To find the BGB with the area for each core:
BGB_core_1_site_C = sum_core_1_site_C/area_plastic_core
print("The BGB for core 1 site C is:", BGB_core_1_site_C, "g m^-2")

BGB_core_2_site_C = sum_core_2_site_C/area_plastic_core
print("The BGB for core 2 site C is:", BGB_core_2_site_C, "g m^-2")

BGB_core_3_site_C = sum_core_3_site_C/area_plastic_core
print("The BGB for core 3 site C is:", BGB_core_3_site_C, "g m^-2")

# Area of metal core for core 4:
BGB_core_4_site_C = sum_core_4_site_C/area_metal_core
print("The BGB for core 4 site C is:", BGB_core_4_site_C, "g m^-2")



                              # TO FIND THE SAMPLE STANDARD DEVIATION OF EACH SITES
    
# SITE A - To find the standard deviation of the 3 cores:
std_site_A_2022 = statistics.stdev([BGB_core_1_site_A, BGB_core_2_site_A, BGB_core_3_site_A])
print("/nThe standard deviation of the site A 2022 is:", std_site_A_2022, "g m$^{-2}$")

# SITE B - To find the standard deviation of the 4 cores:
std_site_B_2022 = statistics.stdev([BGB_core_1_site_B, BGB_core_2_site_B, BGB_core_3_site_B, BGB_core_4_site_B])
print("/nThe standard deviation of the site B 2022 is:", std_site_B_2022, "g m$^{-2}$")

# SITE C - To find the standard deviation of the 4 cores:
std_site_C_2022 = statistics.stdev([BGB_core_1_site_C, BGB_core_2_site_C, BGB_core_3_site_C, BGB_core_4_site_C])
print("/nThe standard deviation of the site C 2022 is:", std_site_C_2022, "g m$^{-2}$")



                                      # FOR THE TOTAL BGB  PER SITES

# SITE A - To have the BGB total of the 3 cores:
total_site_A_2022 = (BGB_core_1_site_A+BGB_core_2_site_A+BGB_core_3_site_A)/3
print("The total of the site A for 2022 is", total_site_A_2022, "g m^-2")



# SITE B - To have the BGB total of the 3 cores of site A:
total_site_B_2022 = (BGB_core_1_site_B+BGB_core_2_site_B+BGB_core_3_site_B+BGB_core_4_site_B)/4
print("The total of the site B for 2022 is", total_site_B_2022, "g m^-2")



# SITE C - To have the BGB total of the 3 cores:
total_site_C_2022 = (BGB_core_1_site_C+BGB_core_2_site_C+BGB_core_3_site_C+BGB_core_4_site_C)/4
print("The total of the site C for 2022 is", total_site_C_2022, "g m^-2")



                                      # TO PREPARE THE DATA TO PLOT

# To prepare the data for the plot:
x = ['Restored Edge', 'Restored Inland', 'Natural']
y = [total_site_A_2022, total_site_B_2022, total_site_C_2022]

# For the error bars:
e = [std_site_A_2022, std_site_B_2022, std_site_C_2022] 



                                        # TO PLOT ON A BAR GRAPH


# To plot a bar graph:
plt.figure()
plt.bar(x, y, color =['pink','linen','paleturquoise'])
plt.gca().invert_yaxis()
plt.title("Total Belowground Biomass of $\mathbf{Avicennia}$ $\mathbf{germinans}$ (g m$^{-2}$) in West Belle Island, LA,\n"
          "Before the December 2022 Mild Freezing Event In Different Sites.", size = '18', fontweight='bold', y=-0.12)
plt.gca().xaxis.set_label_position('top')
plt.gca().xaxis.tick_top()
plt.xlabel("Sites In West Belle Barrier Island, Louisiana.", size = '15', labelpad=10)
plt.xticks(size = '12')
plt.ylabel("Total Belowground Biomass (g m$^{-2}$) of $\it{Avicennia}$ $\it{germinans}$.", size = '15')
plt.yticks(size = '12')
plt.ylim(top=0, bottom=9500)   


# For the standard deviation:
plt.errorbar(x, y, e, linestyle = 'None', marker = '_', color = 'black')

# For the size of the graph:
plt.rcParams['figure.figsize'] = [10, 9]
plt.subplots_adjust()

        
                                                # TO SAVE THE GRAPH ON COMPUTER


# To save as PNG:
#plt.savefig('Plot_of_BGB_December_2022.png', bbox_inches='tight')




#################################################################################################################################################################################################################################################




                                            # FOR THE DATA OF 2023
    
                                         # TO IMPORT THE DATA FROM CSV

# To import data: (Note, I took out the index -- that was making it treat the first column differently)
data2 = pd.read_csv('/Users/agathe/Library/CloudStorage/OneDrive-UniversityofNewOrleans/Foster-Martinez_Coastal_Interactions_Lab_shared/Agathe_Desthomas_Progress_Reports/Python/Total_2022_2023_Combined/BGB_Per_Site_March_2023.csv', header=0)
df2 = pd.DataFrame(data2)



                                        # TO CHANGE THE FONT STYLE OF THE PLOT

# For the plot font:
font = FontProperties()
font.set_family('serif')
font.set_name('Times New Roman')



                            # TO GET RID OF THE NaN VALUES (EMPTY CELLS) FROM CSV FILE

# To get rid of the NaN values (empty cells in csv/excel) and replace them by 0.000:
dff = df2.fillna(0.000)



                                        # TO GET THE SUM OF EACH CORE

# SITE A - To sum core by core:
core_1_site_A_2023 = dff.iloc[0:,1]
sum_core_1_site_A_2023 = sum(core_1_site_A_2023)
print("The sum of the site A core 1 is", sum_core_1_site_A_2023, "g")

core_2_site_A_2023 = dff.iloc[0:,2]
sum_core_2_site_A_2023 = sum(core_2_site_A_2023)
print("The sum of the site A core 2 is", sum_core_2_site_A_2023, "g")

core_3_site_A_2023 = dff.iloc[0:,3]
sum_core_3_site_A_2023 = sum(core_3_site_A_2023)
print("The sum of the site A of core 3 is", sum_core_3_site_A_2023, "g")



# SITE B - To sum core by core:
core_1_site_B_2023 = dff.iloc[0:,4]
sum_core_1_site_B_2023 = sum(core_1_site_B_2023)
print("The sum of the site B core 1 is", sum_core_1_site_B_2023, "g")

core_2_site_B_2023 = dff.iloc[0:,5]
sum_core_2_site_B_2023 = sum(core_2_site_B_2023)
print("The sum of the site B core 2 is", sum_core_2_site_B_2023, "g")

core_3_site_B_2023 = dff.iloc[0:,6]
sum_core_3_site_B_2023 = sum(core_3_site_B_2023)
print("The sum of the site B of core 3 is", sum_core_3_site_B_2023, "g")

core_4_site_B_2023 = dff.iloc[0:,7]
sum_core_4_site_B_2023 = sum(core_4_site_B_2023)
print("The sum of the site B core 4 is:", sum_core_4_site_B_2023, "g")



# SITE C - To sum core by core:
core_1_site_C_2023 = dff.iloc[0:,8]
sum_core_1_site_C_2023 = sum(core_1_site_C_2023)
print("The sum of the site C core 1 is", sum_core_1_site_C_2023, "g")

core_2_site_C_2023 = dff.iloc[0:,9]
sum_core_2_site_C_2023 = sum(core_2_site_C_2023)
print("The sum of the site C core 2 is", sum_core_2_site_C_2023, "g")

core_3_site_C_2023 = dff.iloc[0:,10]
sum_core_3_site_C_2023 = sum(core_3_site_C_2023)
print("The sum of the site C of core 3 is", sum_core_3_site_C_2023, "g")

core_4_site_C_2023 = dff.iloc[0:,11]
sum_core_4_site_C_2023 = sum(core_4_site_C_2023)
print("The sum of the site C core 4 is", sum_core_4_site_C_2023, "g")






                      # TO FIND THE BGB BY TAKING THE SUM AND DIVIDING IT BY THE AREA OF THE CORE

# SITE A - To find the BGB with the area for each core:
BGB_core_1_site_A_2023 = sum_core_1_site_A_2023/area_plastic_core
print("The BGB for core 1 site A is:", BGB_core_1_site_A_2023, "g.m^-2")

BGB_core_3_site_A_2023 = sum_core_3_site_A_2023/area_plastic_core
print("The BGB for core 3 site A is:", BGB_core_3_site_A_2023, "g.m^-2")


# To find the BGB with the area for the core of 0.26m:
BGB_core_2_site_A_2023 = sum_core_2_site_A_2023/area_plastic_core
print("The BGB for core 2 site A is:", BGB_core_2_site_A_2023, "g.m^-2")



# SITE B - To find the BGB with the area for each core:
BGB_core_1_site_B_2023 = sum_core_1_site_B_2023/area_plastic_core
print("The BGB for core 1 site B is:", BGB_core_1_site_B_2023, "g.m^-2")

BGB_core_2_site_B_2023 = sum_core_2_site_B_2023/area_plastic_core
print("The BGB for core 2 site B is:", BGB_core_2_site_B_2023, "g.m^-2")

BGB_core_3_site_B_2023 = sum_core_3_site_B_2023/area_plastic_core
print("The BGB for core 3 site B is:", BGB_core_3_site_B_2023, "g.m^-2")

# Area of metal core for core 4:
BGB_core_4_site_B_2023 = sum_core_4_site_B_2023/area_metal_core
print("The BGB for core 4 site B is:", BGB_core_4_site_B_2023, "g.m^-2")



# SITE C - To find the BGB with the area for each core:
BGB_core_1_site_C_2023 = sum_core_1_site_C_2023/area_plastic_core
print("The BGB for core 1 site C is:", BGB_core_1_site_C_2023, "g.m^-2")

BGB_core_2_site_C_2023 = sum_core_2_site_C_2023/area_plastic_core
print("The BGB for core 2 site C is:", BGB_core_2_site_C_2023, "g.m^-2")

BGB_core_3_site_C_2023 = sum_core_3_site_C_2023/area_plastic_core
print("The BGB for core 3 site C is:", BGB_core_3_site_C_2023, "g.m^-2")

# Area of metal core for core 4:
BGB_core_4_site_C_2023 = sum_core_4_site_C_2023/area_metal_core
print("The BGB for core 4 site C is:", BGB_core_4_site_C_2023, "g.m^-2")




                                # TO FIND THE SAMPLE STANDARD DEVIATION OF EACH SITES

# SITE A - To find the standard deviation of the 3 cores:
std_site_A_2023 = statistics.stdev([BGB_core_1_site_A_2023, BGB_core_2_site_A_2023, BGB_core_3_site_A_2023])
print("The standard deviation of the site A 2023 is:", std_site_A_2023, "g.m^-2")

# SITE B - To find the standard deviation of the 4 cores:
std_site_B_2023 = statistics.stdev([BGB_core_1_site_B_2023, BGB_core_2_site_B_2023, BGB_core_3_site_B_2023, BGB_core_4_site_B_2023])
print("The standard deviation of the site B 2023 is:", std_site_B_2023, "g.m^-2")
        
# SITE C - To find the standard deviation of the 3 cores:
std_site_C_2023 = statistics.stdev([BGB_core_1_site_C_2023, BGB_core_2_site_C_2023, BGB_core_3_site_C_2023, BGB_core_4_site_C_2023])
print("The standard deviation of the site C 2023 is:", std_site_C_2023, "g.m^-2")



                                      # FOR THE TOTAL BGB  PER SITES

# SITE A - To have the BGB total of the 3 cores:
total_site_A_2023 = (BGB_core_1_site_A_2023+BGB_core_2_site_A_2023+BGB_core_3_site_A_2023)/3
print("The total of the site A for 2023 is", total_site_A_2023, "g.m^-2")

# SITE B - To have the BGB total of the 3 cores of site A:
total_site_B_2023 = (BGB_core_1_site_B_2023+BGB_core_2_site_B_2023+BGB_core_3_site_B_2023+BGB_core_4_site_B_2023)/4
print("The total of the site B for 2023 is", total_site_B_2023, "g.m^-2")

# SITE C - To have the BGB total of the 3 cores:
total_site_C_2023 = (BGB_core_1_site_C_2023+BGB_core_2_site_C_2023+BGB_core_3_site_C_2023+BGB_core_4_site_C_2023)/4
print("The total of the site C for 2023 is", total_site_C_2023, "g.m^-2")





##########################################################################################################################################################################


                                      # TO PREPARE THE DATA TO PLOT

# To prepare the data for the plot:
a = ['Restored Edge', 'Restored Inland', 'Natural']
b = [total_site_A_2023, total_site_B_2023, total_site_C_2023]

# For the error bars:
c = [std_site_A_2023, std_site_B_2023, std_site_C_2023]



                                        # TO PLOT ON A BAR GRAPH

# To plot a bar graph:
plt.figure()
plt.bar(a, b, color =['rosybrown','saddlebrown','darkcyan'])
plt.gca().invert_yaxis()
plt.title("Total Belowground Biomass of $\mathbf{Avicennia}$ $\mathbf{germinans}$ (g m$^{-2}$) in West Belle Island, LA,\n"
          "After the December 2022 Mild Freezing Event In Different Sites.", size = '18', fontweight='bold', y=-0.12)
plt.gca().xaxis.set_label_position('top')
plt.gca().xaxis.tick_top()
plt.xlabel("Sites In West Belle Barrier Island, Louisiana.", size = '15', labelpad=10)
plt.xticks(size = '12')
plt.ylabel("Total Belowground Biomass (g m$^{-2}$) of $\it{Avicennia}$ $\it{germinans}$.", size = '15')
plt.yticks(size = '12')
plt.ylim(top=0, bottom=9500)   

# For the standard deviation:
plt.errorbar(a, b, c, linestyle = 'None', marker = '_', color = 'black')

# For the size of the graph:
plt.rcParams['figure.figsize'] = [10, 9]
plt.subplots_adjust()



                                                # TO SAVE THE GRAPH ON COMPUTER


# To save as PNG:
#plt.savefig('Plot_of_BGB_March_2023.png', bbox_inches='tight')






                              # TO PLOT THE TWO PREVIOUS GRAPHS TOGETHER ON THE SAME FIGURE
    
                                            # TO PREPARE THE DATA TO PLOT
# For the columns:
i = ['Restored\nEdge 2022', 'Restored\nEdge 2023', 'Restored\nInland 2022', 'Restored\nInland 2023', 'Natural\n2022', 'Natural\n2023']
j = [total_site_A_2022, total_site_A_2023, total_site_B_2022, total_site_B_2023, total_site_C_2022, total_site_C_2023]


# For the standard deviation:
k = [std_site_A_2022, std_site_A_2023, std_site_B_2022, std_site_B_2023, std_site_C_2022, std_site_C_2023]







                                                # TO PLOT ON A BAR GRAPH
plt.figure()
plt.bar(i, j, color=['gainsboro', 'darkgray', 'silver', 'dimgray', 'lightgray', 'grey'])
plt.gca().invert_yaxis()
#plt.title("Total Belowground Biomass of $\mathbf{Avicennia}$ $\mathbf{germinans}$ (g m$^{-2}$) in West Belle Island, LA,\n" "Before and After The December 2022 Mild Freezing Event in Different Sites.", size='18',fontweight='bold',  y=-0.12)
plt.gca().xaxis.set_label_position('top')
plt.gca().xaxis.tick_top()
plt.xlabel("Sites in West Belle Barrier Island, Lousiana.", size = '15', labelpad=10)
plt.xticks(size='12')
plt.ylabel("Total Belowground Biomass (g m$^{-2}$) of $\it{Avicennia}$ $\it{germinans}$.", size='15')
plt.yticks(size='12')
plt.ylim(top=0, bottom=12000)

# For the standard deviation:
plt.errorbar(i, j, k, linestyle='None', marker='_', color='black')


# For the statistical differences (p-value):

# Before and after site A:
plt.gca().annotate('N.S.', xy=(0.485, 1700), xytext=(0.485, 1750),
            fontsize=10, ha='center', va='top', xycoords='data', 
            arrowprops=dict(arrowstyle='-[, widthB=4.2, lengthB=0.5', lw=1.0))

# Before and after site B:
plt.gca().annotate('N.S.', xy=(2.520, 3100), xytext=(2.520, 3150),
            fontsize=10, ha='center', va='top', xycoords='data', 
            arrowprops=dict(arrowstyle='-[, widthB=4.2, lengthB=0.5', lw=1.0))

# Before and after site C:
plt.gca().annotate('***', xy=(4.525, 9600), xytext=(4.525, 9650),
            fontsize=14, ha='center', va='top', xycoords='data', 
            arrowprops=dict(arrowstyle='-[, widthB=3.05, lengthB=0.5', lw=1.0))




# 2022 Before site A vs B:
plt.gca().annotate('*', xy=(0.990, 3100), xytext=(0.990, 3150),
            fontsize=14, ha='center', va='top', xycoords='data', 
            arrowprops=dict(arrowstyle='-[, widthB=6.2, lengthB=0.5', lw=1.0))

# 2022 Before site B vs C:
plt.gca().annotate('*', xy=(3.0, 10300), xytext=(3.0, 10350),
            fontsize=14, ha='center', va='top', xycoords='data', 
            arrowprops=dict(arrowstyle='-[, widthB=6.2, lengthB=0.5', lw=1.0))

# 2022 Before site A vs C:
plt.gca().annotate('*', xy=(2.025, 9600), xytext=(2.025, 9650),
            fontsize=14, ha='center', va='top', xycoords='data', 
            arrowprops=dict(arrowstyle='-[, widthB=12.25, lengthB=0.5', lw=1.0))




# 2023 After site A vs B:
plt.gca().annotate('**', xy=(2.0, 3800), xytext=(2.0, 3850),
            fontsize=14, ha='center', va='top', xycoords='data', 
            arrowprops=dict(arrowstyle='-[, widthB=6.3, lengthB=0.5', lw=1.0))

# 2023 After site B vs C:
plt.gca().annotate('**', xy=(4.01, 10900), xytext=(4.01, 10950),
            fontsize=14, ha='center', va='top', xycoords='data', 
            arrowprops=dict(arrowstyle='-[, widthB=6.3, lengthB=0.5', lw=1.0))

# 2023 After site A vs C:
plt.gca().annotate('**', xy=(3.018, 11500), xytext=(3.018, 11550),
            fontsize=14, ha='center', va='top', xycoords='data', 
            arrowprops=dict(arrowstyle='-[, widthB=12.5, lengthB=0.5', lw=1.0))




# For the size of the graph:
plt.rcParams['figure.figsize'] = [10, 9]
plt.subplots_adjust()



                                                # TO SAVE THE GRAPHS ON COMPUTER


# To save as PNG:
plt.savefig('Plot_of_BGB_Combined.png', bbox_inches='tight')
```
